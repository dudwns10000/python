- #### sort_index(axis, ascending=True)

### aggregate(func, axis=0, \*args, \*\*kwargs) 또는 agg(func, axis=0, \*args, \**kwargs)
- DataFrame, Series의 메소드로 집계결과를 다양한 형태로 묶어서 볼때 사용한다.
    - **사용자 정의 집계메소드를 사용할 때도 편리하다.**
- 매개변수
    - func 
        - 집계 함수 지정
            - 함수명/함수리스트 (문자열)
                - 판다스 제공 집계메소드들은 문자열로, 사용자정의 집계함수는 함수 객체로 전달
            - 딕션어리 : {'집계할컬럼' : 집계함수, ... }
                - 컬럼별로 다른 집계를 할 경우
    - axis
        - 0 또는 'index' (기본값): 컬럼 별 집계
        - 1 또는 'columns': 행 별 집계
    - \*args, \**kwargs 
        - 함수에 전달할 매개변수. 
        - 집계함수는 첫번째 매개변수로 Series를 받는다. 그 이외의 매개변수가 있는 경우.

#### agg()

```python
import pandas as pd

flights = pd.read_csv('data/flights.csv')
flights.shape
# series에 적용
# 여러개 통계량: 메소드들을 [리스트]로 묶는다. pandas 제공 메소드들: 문자열 메소드 이름만 제공
flights['ARR_DELAY'].agg(['mean','max','min'])
```

```python
# DataFrame
flights.agg(['min','max']).T
```

```python
# 컬럼별로 다른 통계량을 조회할 경우
# flights['ARR_DELAY'].mena(),flights['DEP_DELAY'].sum()
# {컬럼명:통계함수}
flights.agg({
    "ARR_DELAY":"mean",
    "DEP_DELAY":"sum"
})
```

```python
flights.agg({
    "ARR_DELAY":"mean",
    "DEP_DELAY":["sum","mean"],
    "AIRLINE":"mode"
})
```

```python
flights.agg('quantile',q=[0.25,0.5,0.75])
```

# Groupby
- 특정 열을 기준으로 데이터셋을 묶는다.
- ~~ 별 집계를 할 때 사용한다.
- 구문
    - DF.groupby('그룹으로묶을기준컬럼')['집계할 컬럼'].집계함수()
        - groupby에 여러개의 컬럼을 기준으로 나눌 경우 리스트에 묶어서 전달한다.
        - 집계할 컬럼이 여러개인 경우 리스트로 묶어준다.
    - 집계함수
        - 기술통계 함수들
        - agg()/aggregate()
            - 여러 다른 집계함수 호출시(여러 집계를 같이 볼경우)
            - 사용자정의 집계함수 호출시
            - 컬럼별로 다른 집계함수들을 호출할 경우

```python
# AIRLINE 별 도착 딜레이 시간의 평균
flights.groupby('AIRLINE')['ARR_DELAY'].mean().head(5)
```

```python
# AIRLINE별 출발, 도착딜레이 시간의 평균
flights.groupby('AIRLINE')[['DEP_DELAY','ARR_DELAY']].mean().head()
```

```python
# AIRLINE별 출발, 도착딜레이 시간의 최소, 최대값
flights.groupby("AIRLINE")[['DEP_DELAY','ARR_DELAY']].agg(['min','max']).head()
```

```python
# AIRLINE별 출발딜레이-최소,최대값, 도착 딜레이-합계, 평균 (컬럼별로 다른 통계량 조회)
agg_dict = {
    'DEP_DELAY':['min','max'],
    'ARR_DELAY':['sum','mean'],
}
flights.groupby('AIRLINE').agg(agg_dict).head()
```

###  복수열 기준 그룹핑
- 두개 이상의 열을 그룹으로 묶을 수 있다. 
- groupby의 매개변수에 그룹으로 묶을 컬럼들의 이름을 리스트로 전달한다.

```python
# 항공사의 월별 도착 딜레이시간 평균
flights.groupby(['AIRLINE','MONTH'])['ARR_DELAY'].mean().head()
```

```python
# 항공사 요일별 회항/취소 횟수
flights.groupby(['AIRLINE','WEEKDAY'])[['DIVERTED','CANCELLED']].sum().head()
```

## 사용자 정의 집계함수를 만들어 적용

### 사용자 정의 집계 함수 정의
- 매개변수
    1. Series 또는 DataFrame을 받을 매개변수(필수)
    2. 필요한 값을 받을 매개변수를 선언한다. (선택)

### agg() 를 사용해 사용자 정의 집계 함수 호출
- DataFrame.agg(func=None, axis=0, \*args, \*\*kwargs)
    - axis : 사용자 정의 함수에 전달할 값들(Series)의 축 지정
- Series.agg(func=None, axis=0, \*args, \*\*kwargs)
    - DataFrame의 agg와 매개변수 구조를 맞추기 위해 axis 지정한다. (kwargs를 이용해 매개변수 전달할 경우 axis는 생략해도 된다. - axis기본값 0을 그냥 쓰면 되므로. - keyword 인자 뒤에 position 인자는 안되서 *args로 값 전달시에는 axis를 지정해야 한다.)
- DataFrameGroupBy.agg(func, \*args, \*\*kwargs) : 
    - axis 지정안함.
    - 사용자 함수에 Series를 group 별로 전달한다.
- SeriesGroupBy.agg(func=None,  \*args, \*\*kwargs)  
    - axis 지정안함
    - 사용자 함수에 Series를 group 별로 전달한다.
  
- \*args, \*\*kwargs는 사용자 정의 함수에 선언한 매개변수가 있을 경우 전달할 값을 전달한다.  
    - 키워드 인자를 이용해 가변인자로 전달하는 것이 편하다.

#### 매개변수 없는 사용자 정의 함수
- 최대 값과 최소값의 차이를 반환

```python
# 사용자정의 집계함수(통계함수) => 최대값과 최소값의 차이를 구하는 함수
# 매개변수로 Series나 DataFrame을 받을 변수, 그 외에 필요한 변수를 추가 선언
import numpy as np
def max_min_diff(x):
#    x: series, dataframe을 받을 변수
#    x가 수치형(ind,float)이 아니면 예외를 발생
    if x.dtype == 'object':   #np.object -> object타입
        raise TypeError('수치형 타입만')
    return x.max() - x.min()
```

```python
max_min_diff(flights['ARR_DELAY'])
```

```python
# ARR_DELAY의 최소값,최대값,최대값과 최소값의 차이
# 판다스 제공 통계량 함수: 함수이름을 문자열로 제공
# 사용자정의 통계량 함수: 함수객체
flights['ARR_DELAY'].agg(['min','max',max_min_diff])
```
